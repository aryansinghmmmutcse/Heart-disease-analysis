```python
plt.figure(figsize=(7,5))
sns.countplot(d['target'],hue=d['target'])
plt.show()
```

```python
#Chest pain distribution as per target variable
sns.countplot(d['cp'],hue=d['target'],data=d)
plt.legend(labels=['No-disease','Disease'])
plt.xticks([0,1,2,3],['Low','Little','Medium','High'])
plt.show()
```

```python
#compare resting blood pressure as per sex column
g=sns.FacetGrid(d,hue='sex',aspect=4)
g.map(sns.kdeplot,'trestbps')
plt.show()
```

```python
#plot categorical variable
ct=[]
cn=[]
for i in d.columns:
    if d[i].nunique()<=10:
        ct.append(i)
    else:
        cn.append(i)
d.hist(cn)
```
